Task 1 – Analysis on ML Test Scores

Introduction:   
In this assignment, I analyzed the Machine Learning test scores of students belonging to three different batches. The goal of this analysis is to understand the performance of each batch and derive meaningful insights using Python.   

Objective:    
To analyze ML test scores of students from three different batches and derive meaningful insights using Python and Pandas.   

Dataset Description:   
The dataset contains the following columns:    
-> Batch – Name of the batch  
-> User_ID – Unique student identifier   
-> Score – Student score out of 7   

Step 1: Import Required Libraries

I used the pandas library for data loading and analysis.

```python
import pandas as pd
```

Step 2: Loading the Dataset

I loaded the dataset using the read_csv() function.

```python
df = pd.read_csv("scores_data.csv")
df.head()
```

Step 3: Data Cleaning

3.1 Removing Extra Spaces from Column Names

I noticed that some column names contained extra spaces, so I removed them to avoid errors.

```python
df.columns = df.columns.str.strip()
```

3.2 Converting Score Column to Numeric

The Score column was in text format (e.g., "4 / 7").    
To perform numerical analysis, I extracted the numeric value and converted it into an integer.

```python
df['Score_num'] = df['Score'].str.extract(r'(\d+)').astype(int)
```

Step 4: Batch-wise Statistical Analysis

I performed batch-wise analysis to calculate:

-> Number of students  
-> Mean score   
-> Median score    
-> Minimum score    
-> Maximum score    
-> Standard deviation    

```python
summary = df.groupby('Batch')['Score_num'].agg(
    count='count',
    mean='mean',
    median='median',
    min='min',
    max='max',
    std='std'
).round(2)

summary
```
